## Task B

We can start calculating the Shapley value by calculating each player's contribution to a specific coalition:

v() = 0 <br>
v(A) = 20 <br>
v(B) = 20 <br>
v(C) = 60 <br>
v(A,B) = 60 A: +40 B: +40<br>
v(A,C) = 70 A: +10 C: +50<br>
v(B,C) = 70 B: +10 C: +50<br>
v(A,B,C) = 100 A: +30 B: +30 C: +40<br>

Next we can calculate the permutation coefficients with respect to subset size:
$$ |P|! = 6$$
$$ |S| = 0 (|P| - |S| - 1)! = 2! = 2$$
$$ |S| = 1 (|P| - |S| - 1)! = 1! = 1$$
$$ |S| = 2 (|P| - |S| - 1)! = 0! = 1$$
Now we can calculate the Shapley value for player A:
$$\phi_A = \dfrac 16 \left(2 * 20 + 40 + 10 + 30 \right) = 20$$

```python
import pandas as pd
df = pd.read_csv("adult.csv")
df["income"] = pd.get_dummies(df["income"], drop_first = True)
cols=df.select_dtypes(exclude=['int', 'uint8']).columns.to_list()
df[cols] = df[cols].astype("category")

X = df.iloc[:, :-1]
target = df["income"]
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, target_train, target_test = train_test_split(pd.get_dummies(X, drop_first = True), target, random_state = 42, train_size = 0.8, stratify = target)
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
model = xgb.XGBClassifier()
model.fit(X_train, target_train)
accuracy_score(target_test, model.predict(X_test))
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()

logistic_model.fit(X_train, target_train)
accuracy_score(target_test, logistic_model.predict(X_test))
```
